```python
tornado_years['STATE_ABBR'].value_counts()
```

```python
tornado_years = tornado_years.loc[tornado_years['EVENT_TYPE']=='Tornado']
tornado_years['STATE_ABBR'].value_counts()
```

```python
tornado_years = tornado_years.drop(columns=['BEGIN_LOCATION', 'EVENT_TYPE', 'MAGNITUDE', 'DAMAGE_CROPS_NUM', 'CZ_TIMEZONE', 'MAGNITUDE_TYPE', 
                                            'EPISODE_ID', 'CZ_TYPE', 'CZ_FIPS', 'WFO', 'FLOOD_CAUSE', 'BEGIN_RANGE', 'BEGIN_AZIMUTH', 'END_RANGE',
                                            'END_AZIMUTH', 'END_LOCATION','TOR_LENGTH', 'TOR_WIDTH', 'DAMAGE_PROPERTY_NUM',
                                            'INJURIES_INDIRECT', 'DEATHS_INDIRECT'])
```

```python
tornado_years
```

```python
tornado_years[["BEGIN_MONTH", "BEGIN_DAY", "BEGIN_YEAR"]] = tornado_years["BEGIN_DATE"].str.split("/", expand = True)
tornado_years[["END_MONTH", "END_DAY", "END_YEAR"]] = tornado_years["END_DATE"].str.split("/", expand = True)
```

```python
tornado_years
```

```python
tornado_years['BEGIN_TIME'] = tornado_years['BEGIN_TIME'].astype(str)
tornado_years.info()
```

```python
tornado_years = tornado_years.reset_index(drop = True)
```

```python
n = 0
for i in tornado_years['BEGIN_TIME']:
    if len(re.findall(r'(\d)', i)) == 3:
        tornado_years.at[n, 'BEGIN_TIME'] = ((i[:1] + ':' + i[1:]))
    if len(re.findall(r'(\d)', i)) == 4:
        tornado_years.at[n, 'BEGIN_TIME'] = ((i[:2] + ':' + i[2:]))
    n += 1
    
tornado_years
```
