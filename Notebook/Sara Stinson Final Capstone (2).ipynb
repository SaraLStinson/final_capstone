```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import glob
import os
import datetime as dt

%matplotlib inline
```

```python
tor = pd.read_csv('../data/all_tornado.csv')
tor.info()
```

```python
tor['BEGIN_LAT'] = tor['BEGIN_LAT'].replace(' ','')
tor['BEGIN_LAT'] = tor['BEGIN_LAT'].replace('', None)
tor['BEGIN_LON'] = tor['BEGIN_LON'].replace(' ','')
tor['BEGIN_LON'] = tor['BEGIN_LON'].replace('', None)
```

```python
tor['BEGIN_LAT'] = tor['BEGIN_LAT'].astype(float)
tor['BEGIN_LON'] = tor['BEGIN_LON'].astype(float)
```

```python
tor_lat = tor.groupby('BEGIN_YEAR')['BEGIN_LAT'].mean().to_frame('Average Lat by Year').reset_index()
tor_lat
```

```python
tor_lon = tor.groupby('BEGIN_YEAR')['BEGIN_LON'].mean().to_frame('Avg Lon by Year').reset_index()
```

```python
tor_lon
```

```python
lat_lon = pd.merge(tor_lat,tor_lon)
```

```python
lat_lon.info()
```

```python
incr_5 = lat_lon.groupby(lat_lon.index // 5).mean()
```

```python
incr_5
```

```python
incr_5['BEGIN_YEAR'] = ['1950-1954', '1955-1959','1960-1964','1965-1969','1970-1974','1975-1979','1980-1984',
                      '1985-1989','1990-1994','1995-1999','2000-2004','2005-2009','2010-2014','2015-2019','2020-2021']
```

```python
incr_5
```

```python
incr_5.to_csv('Average Lat_Lon by Year Range.csv')
```
